```python
def gini(X_train, X_test, y_train,i):
    model = DecisionTreeClassifier(criterion = "gini",random_state = 10,max_depth=i, min_samples_leaf=5)
    model.fit(X_train, y_train)
    return model
```

```python
def entropy(X_train, X_test, y_train,i):
    model1= DecisionTreeClassifier(criterion = "entropy", random_state = 100,max_depth = i, min_samples_leaf = 5)
    model1.fit(X_train, y_train)
    return model1
```

```python
def prediction(X_test, clf_object):
    y_pred = clf_object.predict(X_test)
    return y_pred
```

```python
def accuracy(y_test, y_pred):
    print ("Accuracy score : ",accuracy_score(y_test,y_pred)*100)
```
